```python
df = pd.read_csv('sales.csv')
```

```python
df['category'] = 'NA'
df
```

```python
df.loc[(df['sales'] < df['sales'].quantile(.25)),'category'] = 'BAJA VENTA'
df.loc[(df['sales'] >= df['sales'].quantile(.25)) & (df['sales'] < df['sales'].quantile(.5)),'category'] = 'MEDIA VENTA'
df.loc[(df['sales'] >= df['sales'].quantile(.25)) & (df['sales'] < df['sales'].quantile(.5)),'category'] = 'MEDIA VENTA'
```

```python
df = pd.read_csv('metrics.csv')
df
```

```python
df_2 = pd.read_csv('sales.csv')

df_2
```

```python
dl = df.apply(lambda x: df_2.loc[df_2['product'] == x['prod_B'],'category'].values, axis=1)

def ungroup(value):
    try:
        [n_value] = value
        return n_value
    except:
        return 'NA'
    
df['category_B'] = dl.apply(lambda x:ungroup(x))

df
```
